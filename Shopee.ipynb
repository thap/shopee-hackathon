```python
import heapq

def heapsort(iterable):
    h = []
    for item in iterable:
        heapq.heappush(h, (-len(item.split()), item))
    return [heapq.heappop(h) for i in range(len(h))]
```

```python
from collections import defaultdict
import heapq
vocab = defaultdict(set)
for idx in keyword.index:
    words = keyword.loc[idx,'Keywords']
    for word in words.split(', '):
        vocab[word].add(idx)
vocab
```

```python
from collections import defaultdict
mis_map = defaultdict(set)
for col in mismatch.columns:
    if col in vocab:
        for word in mismatch[col].dropna():
            mis_map[col].add(word)
print(mis_map)
```

```python
from collections import defaultdict
mis_map = defaultdict(list)
words = []
with open('open-round1/Extra Material 3 - mismatch list.csv', 'r') as f:
    line = f.readline().strip()[1:]
    for word in line.split(','):
        words.append(word)
    line = f.readline().strip()
    while line:
        for i, word in enumerate(line.split(',')):
            if word and words[i] in vocab:
                mis_map[words[i]].append(word)
        line = f.readline().strip()
print(mis_map)
```

```python
%%time
import re
from collections import deque
with open('out.csv', 'w') as f:
    f.write('index,groups_found\n')
    for idx in question.index:
        res = [w.lower() for w in re.findall(r'[\w-]+', question.loc[idx, 'name'])]
        lst = set()
        in_vocab = set()
        for l in range(len(res), 0, -1):
            for i in range(len(res)-l+1):
                word = ' '.join(res[i:i+l])
                if word in vocab:
                    found = False
                    for w in in_vocab:
                        if word in w.split():
                            found = True
                            break
                    if word in mis_map:
                        for mis_word in mis_map[word]:
                            if mis_word in res:
                                found = True
                                break
                    if not found:
                        in_vocab.add(word)
                        lst.update(vocab[word])
        if idx % 100000 == 0:
            print(idx, str(sorted(list(lst))), res)
        f.write(str(idx) + ',"' + str(sorted(list(lst))) + '"\n')
```
