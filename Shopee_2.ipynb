```python
from collections import deque

def DFS(buyer, seller):
    visit = set()
    queue = [[buyer]]
    
    while queue:
        path = queue.pop(0)
        v = path[-1]
        if v == seller:
            return path[:-1]
        elif v not in visit:
            for u in g[v]:
                new_path = list(path)
                new_path.append(u)
                queue.append(new_path)
            visit.add(v)
    return []
        
```

```python
%%time

with open('out2.csv', 'w') as f:
    f.write('orderid,is_fraud\n')
    for idx, order in orders.iterrows():
        if idx % 100000 == 0:
            print(idx)
        oid, buyer, seller = order
        #print(oid, buyer, seller)
        if buyer == seller:
            f.write(str(oid) + ',' + str(buyer) + '\n')
            continue
        if buyer in g and seller in g:
            res = DFS(buyer, seller)
            if res:
                f.write(str(oid) + ',')
                for link in res:
                    if isinstance(link, str):
                        f.write('"' + link + '"->')
                    else:
                        f.write(str(link) + '-')
                f.write(str(seller) + '\n')
            else:
                f.write(str(oid) + ',not fraud\n')
        else:
            f.write(str(oid) + ',not fraud\n')
```
